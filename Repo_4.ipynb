```python
import pandas as pd

# Ganti nama file sesuai dengan nama file kamu
df = pd.read_csv('tsla_raw_data.csv')
df.head()
```

```python
print(df.columns)
```

```python
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)
```

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np

# Ambil nilai harga penutupan (close)
close_prices = df['close'].values.reshape(-1, 1)

# Normalisasi harga
scaler = MinMaxScaler(feature_range=(0,1))
scaled_data = scaler.fit_transform(close_prices)
```

```python
X_train = []
y_train = []

for i in range(60, len(scaled_data)):
    X_train.append(scaled_data[i-60:i, 0])
    y_train.append(scaled_data[i, 0])

X_train, y_train = np.array(X_train), np.array(y_train)

# Bentuk array ke format [samples, timesteps, features]
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# Buat model
model = Sequential()

# LSTM layer pertama (return_sequences=True agar bisa diteruskan ke layer LSTM berikutnya)
model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], 1)))
model.add(Dropout(0.2))

# LSTM layer kedua
model.add(LSTM(units=50, return_sequences=False))
model.add(Dropout(0.2))

# Dense layer untuk output akhir (1 nilai prediksi)
model.add(Dense(units=1))

# Kompilasi model
model.compile(optimizer='adam', loss='mean_squared_error')

# Latih model
model.fit(X_train, y_train, epochs=25, batch_size=32)
```

```python
# Ambil 60 hari terakhir dari data untuk prediksi
test_data = scaled_data[-80:]  # ambil lebih banyak kalau perlu

X_test = []
for i in range(60, len(test_data)):
    X_test.append(test_data[i-60:i, 0])

X_test = np.array(X_test)
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))
```

```python
# Prediksi harga saham menggunakan model yang sudah dilatih
predicted_prices = model.predict(X_test)

# Balikkan hasil normalisasi ke nilai harga asli
predicted_prices = scaler.inverse_transform(predicted_prices)
```

```python
# Ambil harga asli dari data yang sesuai
actual_prices = close_prices[-len(predicted_prices):]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.plot(actual_prices, color='blue', label='Actual TSLA Price')
plt.plot(predicted_prices, color='red', label='Predicted TSLA Price')
plt.title('TSLA Stock Price Prediction')
plt.xlabel('Time')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

mse = mean_squared_error(actual_prices, predicted_prices)
rmse = np.sqrt(mse)
mae = mean_absolute_error(actual_prices, predicted_prices)

print(f"Mean Absolute Error (MAE): {mae}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
```
